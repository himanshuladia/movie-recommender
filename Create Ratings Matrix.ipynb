```python
import pandas as pd
import numpy as np
```

```python
links = pd.read_csv('ml-latest-small/links.csv', dtype={'imdbId':str, 'tmdbId':str})
ratings = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
df = pd.merge(ratings, links, on='movieId')
```

```python
df.drop(['movieId', 'timestamp', 'tmdbId'], axis=1, inplace=True)
```

```python
df.head()
```

```python
distinct_users = df['userId'].unique()
```

```python
distinct_movies = df['imdbId'].unique()
```

```python
database = np.zeros((len(distinct_users), len(distinct_movies)))
```

```python
database = pd.DataFrame(database, index=distinct_users, columns=distinct_movies)
```

```python
database.head()
```

```python
for user in distinct_users:
    userdf = df[df['userId']==user]
    usermovies = userdf['imdbId']
    for movie in usermovies:
        database.loc[user, movie] = np.float64(userdf[userdf['imdbId']==movie]['rating'])
```

```python
database.to_csv('database.csv', index=False)
```
